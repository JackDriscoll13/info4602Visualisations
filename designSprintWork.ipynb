```python
import altair as alt
import pandas as pd 
import numpy as np
```

```python
majordf = pd.read_csv("collegemajor.csv")

# Read in data and convert enemployment ratio to a % and a string
majordf['Unemployment_rate']=majordf['Unemployment_rate'].apply(lambda x: x*100).apply(lambda x: str(x)+'%')
majordf.head(10)
```

```python
chart = alt.Chart(majordf).mark_point().encode(
    x='Total',
    y='Median',
    color = 'Major_category',
    tooltip =['Major','Median','Unemployment_rate']
).interactive()

chart
```

```python
majordf['%women'] = majordf['Women']/majordf['Total']
bycategorydf = majordf.groupby('Major_category').agg({'Median':np.mean,'%women':np.mean}).reset_index()
bycategorydf.head(10)
```

```python
chart = alt.Chart(bycategorydf).mark_bar().encode(
    x='Major_category',
    y='Median',
    color = '%women',
    tooltip = '%women'                      
).interactive()
chart
```
